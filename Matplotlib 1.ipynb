```python
# * in red color
```

```python
xpoints=np.array([3,9,27])
ypoints=np.array([1,3,9])
plt.plot(xpoints,ypoints,"*"'r')
plt.show()
```

```python
# color for line color and linestyle is linetype

# ms = markersize
```

```python
xpoints=np.array([3,9,27])
ypoints=np.array([1,5,2])
plt.plot(xpoints,ypoints,marker='+',ms=10,linestyle='dashed',color="green")
plt.show()
```

```python
# ms=marker_size mec=marker_border_color k=black
```

```python
xpoints=np.array([3,9,27])
ypoints=np.array([1,5,2])
plt.plot(xpoints,ypoints,linestyle='dotted',marker="*",ms=10,mec="k",color="orange")
plt.show()
```

```python
# ms=marker_size mec=marker_border_color color=linecolor marker=point_style
```

```python
xpoints=np.array([3,9,27])
ypoints=np.array([1,7,9])
plt.plot(xpoints,ypoints,linestyle='dotted',marker="o",ms=30,mec="red",color="orange")
plt.show()
```

```python
# mfc= marker_inside_color or face_color,  mec=marker_border_color , color=linecolor
```

```python
xpoints=np.array([3,9,27])
ypoints=np.array([5,3,9])
plt.plot(xpoints,ypoints,linestyle='dotted',linewidth=5,marker="*",ms=20,mec="red",mfc="skyblue",color="orange")
plt.show()
```
