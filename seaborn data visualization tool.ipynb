```python
sns.distplot(df["length"],color='hotpink')
plt.title("My graph",color='blue',fontsize=25)
plt.show()
```

```python
sns.distplot(df["length"],color='hotpink')
plt.title("My graph",color='blue',size=50)
plt.xlabel("longness")
plt.ylabel("thickness")
plt.show()
```

```python
sns.distplot(df["length"],color='hotpink', vertical=True, hist=False)
plt.title("My graph",color='blue',size=30)
plt.ylabel("longness",loc="top",color="orange")
plt.xlabel("thickness",color="brown",loc="right")
plt.show()
```

```python
sns.distplot(df["length"],color='maroon',kde=True)
plt.title("My graph",color='blue',size=30)
plt.xlabel("longness",loc="right",color="orange")
plt.ylabel("thickness",color="brown",loc="top")
plt.show()
```
